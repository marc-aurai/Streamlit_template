```python
import requests
import datetime as dt
import pandas as pd
```

# Get all dates from competition:
d1k1pqdg2yvw8e8my74yvrdw4 = 2022/2023

```python
url = f'http://api.performfeeds.com/soccerdata/tournamentschedule/6bxvue6su4ev1690mzy62a41t/d1k1pqdg2yvw8e8my74yvrdw4?_rt=b&_fmt=json'

response = requests.get(url)
match_dates = response.json()['matchDate']
```

```python
len(match_dates)
```

```python
df_matches = pd.DataFrame()
for idx_match_date in range(len(match_dates)):
    df_matches = pd.concat([df_matches, pd.DataFrame(match_dates[idx_match_date]['match'])])
df_matches_ = df_matches[['id','date','homeContestantId','awayContestantId',
                          'homeContestantOfficialName', 'awayContestantOfficialName',
                          'homeContestantName', 'awayContestantName','homeContestantCode','awayContestantCode',
                          ]]
```

```python
df_matches.sort_values(by='date', ascending = False, inplace = True)
display(df_matches[70:90])
```

# Perform merge on ESPN Scraped data

```python
df_espn = pd.read_csv('../../espn_scraper/scraper_data/articles_1500.csv', sep=";")
df_espn.sort_values(by='date', ascending = False, inplace = True)

df_merged = pd.merge(df_espn, df_matches_, how='left', left_on=['date', 'home_abbrev', 'away_abbrev'], right_on=['date', 'homeContestantCode', 'awayContestantCode'])
```

# Get all Eredivisie ID's / Year / Startdate

```python
competetions = requests.get(f'https://api.performfeeds.com/soccerdata/tournamentcalendar/6bxvue6su4ev1690mzy62a41t?_rt=b&_fmt=json')
print(competetions.json().keys())
```

```python
for competetion in competetions.json()['competition']:
    if competetion['name'] == "Eredivisie":
        for eredivisie in competetion['tournamentCalendar']:
            print(eredivisie['id'], eredivisie['name'], eredivisie['startDate'], eredivisie['endDate'])
```
